```python
def clean_text(text):
    cleaned_text = re.sub(r'[^\d\s]', '', str(text))  # Keep only digits and white spaces
    return cleaned_text.strip()  # Remove leading and trailing spaces (optional)

# Apply the cleaning function to the specified column
df["extracted_text"] = df["extracted_text"].apply(clean_text)
```

```python
# Drop duplicates based on img_id, keeping only the last occurrence
df = df.drop_duplicates(subset='img_id', keep='last')
```

```python
df.head(10)
```

```python
def process_text(text):
    split_text = text.split(" ")
    numbers = [part for part in split_text if part.isdigit() or part.replace('٫', '').isdigit() or part.replace(',', '').isdigit()]
    if len(numbers) >= 2:
        first_number = numbers[0]
        second_number = numbers[1]
        
        # Swap if the length of the first number is greater than the second
        if len(first_number) >= len(second_number):
            first_number, second_number = second_number, first_number
        
        # Remove the white space between the numbers
        return first_number + second_number
    else:
        return text

# Apply the process function to the extracted_text column
df['extracted_text'] = df['extracted_text'].apply(process_text)
```

```python
# Extract base image ID (ignoring the last part)
df['base_img_id'] = df['img_id'].apply(lambda x: '_'.join(x.split('_')[:-1]))

# Drop duplicates based on the base image ID, keeping only the last occurrence
df = df.drop_duplicates(subset='base_img_id', keep='last')

# Drop the base_img_id column as it's no longer needed
df = df.drop(columns=['base_img_id'])
```

```python
# Extract the numerical part from the image ID for sorting
df['img_num'] = df['img_id'].apply(lambda x: int(x.split('_')[0]))

# Sort by the numerical part of the image ID
df = df.sort_values(by='img_num').reset_index(drop=True)

# Drop the temporary column used for sorting
df = df.drop(columns=['img_num'])
```
